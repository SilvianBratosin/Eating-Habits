```python
import pandas as pd

dataset=pd.read_excel('UDJG_CHILD_CONS.xls')
dataset2=pd.read_excel('UDJG_CHILD_FOODS.xls')
dataset3=pd.read_excel('UDJG_CHILD_SUBJ.xls')
```

```python
columns_to_delete=['SURVEY', 'CONDAY', 'CONMONTH', 'CONYEAR', 'EXCEPTIONDAY', 'TIMEHOUR', 'TIMEMINUTES', 'UNITMEAS', 'BRAND', 'ORFACETSCODE', 'ORFACETS', 'ENFACETS', 'PACKAGE', 'OFACETS', 'ORRECIPECODE', 'ENRECIPEDESC', 'AMOUNTRECIPE', 'FPREP']
dataset=dataset.drop(columns=columns_to_delete, axis=1)
dataset.head(10)
```

```python
columns_to_delete2=['SURVEY', 'COMMENTSFOOD', 'ENERGY', 'WATER', 'FAT', 'CARB', 'PROTEINS', 'ALCOHOL', 'ENFOODNAME']
dataset2=dataset2.drop(columns=columns_to_delete2, axis=1)
dataset2.head(10)
```

```python
columns_to_delete3=['SURVEY', 'COUNTRY', 'BIRTHDAY', 'BIRTHMONTH', 'BIRTHYEAR', 'FANTYEAR', 'SWEIGHT', 'SHEIGHT', 'SANTDAY', 'SANTMONTH', 'SANTYEAR', 'MWEIGHT', 'MHEIGHT', 'ENRGYINTAKE', 'UNOVREP', 'WF', 'SPECIALCON', 'SPECDIET', 'ACTIVITY', 'ETHNIC', 'FANTMONTH', 'FANTDAY']
dataset3=dataset3.drop(columns=columns_to_delete3, axis=1)
dataset3.head(10)
```

```python
for i, j in dataset.iterrows():
  #if pd.isna(j['ORRECIPEDESC']) or pd.isna(j['FOODEXRCODE']):
  #if j.isnull().any():
  for i2, j2 in dataset2.iterrows():
    if j['ORFOODCODE']==j2['ORFOODCODE']:
      dataset.at[i, 'FOODEXRCODE'] = dataset2.at[i2, 'FOODEXCODE']
      dataset.at[i, 'ORRECIPEDESC'] = dataset2.at[i2, 'ORFOODNAME']
```

```python
dataset.head()
```

```python
dataset.insert(loc=12, column='GENDER', value='')
dataset.insert(loc=13, column='AGE', value='')
dataset.insert(loc=14, column='WEIGHT', value='')
dataset.insert(loc=15, column='HEIGHT', value='')
dataset.insert(loc=16, column='GEO', value='')
```

```python
for i, j in dataset.iterrows():
  for i2, j2 in dataset3.iterrows():
    if j['ORSUBCODE']==j2['ORSUBCODE']:
      dataset.at[i, 'GENDER'] = dataset3.at[i2, 'GENDER']
      dataset.at[i, 'AGE'] = dataset3.at[i2, 'AGE']
      dataset.at[i, 'WEIGHT'] = dataset3.at[i2, 'WEIGHT']
      dataset.at[i, 'HEIGHT'] = dataset3.at[i2, 'HEIGHT']
      dataset.at[i, 'GEO'] = dataset3.at[i2, 'GEO']
```

```python
dataset=dataset.sample(frac=1, random_state=42)
dataset.reset_index(drop=True, inplace=True)
dataset.head(20)
```

```python
dataset.to_csv('ProcessedData.csv', index=False, header=True, sep=';')
```
